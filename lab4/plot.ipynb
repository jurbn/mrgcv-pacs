```python
times_dict = {'2,2': [2510, 2466, 2450, 2481, 2509], '2,4': [1677, 1671, 1667, 1694, 1660], '4,2': [1875, 1880, 1873, 1906, 1989], 
              '16,16': [1579, 1573, 1596, 1597, 1606], '1,1': [7710, 7706, 7740, 7706, 7836], '32,32': [1590, 1544, 1606, 1569, 1577],
              '4,4': [1745, 1735, 1704, 1869, 1704], '256,192': [1571,1584,1603,1625,1628], '8,8': [1620, 1647, 1643, 1617, 1709],
              '8,1': [1704, 1691, 1678, 1718, 1775], '1,8': [1625, 1621, 1664, 1617, 1686], '1,16': [1646, 1758, 1686, 1649, 1690],
              '16,1': [1717, 1632, 1648, 1619, 1642]}
```

```python
# create a new dict with '1,1', '2,2', '4,4', '8,8', '16,16', '32,32' as keys
plot_dict = {}
for key in ['1,1', '2,2', '4,4', '8,8', '16,16', '32,32']:
    plot_dict[key] = times_dict[key]
```

```python
import numpy as np
# average the values for each key
for key in plot_dict.keys():
    plot_dict[key] = np.mean(plot_dict[key])
```

```python
# plot the values (add the value on every point)
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(plot_dict.keys(), plot_dict.values(), 'o-')
plt.xlabel('Sections')
plt.ylabel('Time (ms)')
plt.title('Time taken for different regions')
plt.show()
```

```python
# print the average of values for times_dict['2,4'] and times_dict['4,2'] and times_dict['1,8'] and times_dict['8,1']
print('Average time for 2x4 regions: ', np.mean(times_dict['2,4']))
print('Average time for 4x2 regions: ', np.mean(times_dict['4,2']))
print('Average time for 1x8 regions: ', np.mean(times_dict['1,8']))
print('Average time for 8x1 regions: ', np.mean(times_dict['8,1']))
```
