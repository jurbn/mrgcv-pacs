```python
import pandas as pd
```

```python
histogram_data = pd.read_csv('histogram.csv', sep=',', header=None)
histogram_data
```

```python
# plot the data with matplotlib
import matplotlib.pyplot as plt
import numpy as np
red = histogram_data[0]
green = histogram_data[1]
blue = histogram_data[2]
```

```python
# the histogram of the data
plt.figure(figsize=(10, 5))
plt.hist(red, 256, density=True, facecolor='red', alpha=0.75)
plt.hist(green, 256, density=True, facecolor='green', alpha=0.75)
plt.hist(blue, 256, density=True, facecolor='blue', alpha=0.75)
plt.xlabel('Red')
plt.ylabel('Probability')
plt.title('Histogram of Red')
plt.grid(True)
```

```python
# do a line plot of the data
plt.figure(figsize=(10, 5))
plt.plot(red, color='red')
plt.plot(green, color='green')
plt.plot(blue, color='blue')
plt.xlabel('Intensity')
plt.ylabel('Number')
plt.title('Intensity of RGB')
plt.grid(True)
plt.show()
```
