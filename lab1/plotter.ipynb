```python
import pandas as pd

# load the csv data into pandas dataframes
basic_data = pd.read_csv('benchmark_basic.csv')
optimized_data = pd.read_csv('benchmark_optimized.csv')
eigen_data = pd.read_csv('benchmark_vectorized.csv')
eigen_not_vectorized_data = pd.read_csv('benchmark_not_vectorized.csv')

# combine the dataframes using column 0 as the index and use just the column 1 of the rest of the dataframes
basic_data = basic_data.set_index(basic_data.columns[0])
optimized_data = optimized_data.set_index(optimized_data.columns[0])
eigen_data = eigen_data.set_index(eigen_data.columns[0])
eigen_not_vectorized_data = eigen_not_vectorized_data.set_index(eigen_not_vectorized_data.columns[0])

data = pd.concat([basic_data, optimized_data, eigen_data, eigen_not_vectorized_data], axis=1)

#plot the data in a linear scale
ax = data.plot(figsize=(10, 10), title='Matrix multiplication benchmark')
ax.set_xlabel('Matrix size (N)')
ax.set_ylabel('Time (ms)')
ax.grid()
ax.legend(['Basic', 'Optimized', 'Eigen', 'Eigen not vectorized'])
ax.set_xlim(100, 10000)
# plot the data in a logarithmic scale (just y axis is in logaritmic scale)
ax = data.plot(logy=True, figsize=(10,10), grid=True, title='Matrix multiplication benchmark')
ax.set_xlabel('Matrix size (N)')
ax.set_ylabel('Time (ms)')
ax.legend(['Basic', 'Optimized', 'Eigen', 'Eigen not vectorized'])
ax.set_xlim(100, 10000)
# plot the data in a logaritmic scale (both axis are in logaritmic scale)
ax = data.plot(loglog=True, figsize=(10,10), grid=True, title='Matrix multiplication benchmark')
ax.set_xlabel('Matrix size (N)')
ax.set_ylabel('Time (ms)')
ax.legend(['Basic', 'Optimized', 'Eigen', 'Eigen not vectorized'])
ax.set_xlim(100, 10000)
```
